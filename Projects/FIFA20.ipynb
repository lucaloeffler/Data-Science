```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Import the Dataset
```

```python
fifa = pd.read_csv("players_20.csv")
```

```python
fifa.info()
#Some information
```

```python
fifa.columns
```

## Which foot do footballers prefer?

```python
fifa_Foot=fifa.groupby('preferred_foot').count()
listA=list(fifa_Foot.index.values)
explode = (0.1, 0)
plt.pie(fifa_Foot.age,explode=explode,autopct='%1.1f%%', labels=listA)
plt.axis('equal')
plt.title("R or L")
```

## The best young player?

```python
dataYoung=fifa[fifa.age<=20].sort_values(['potential','value_eur'],ascending = [False,True]).head(10)
sns.barplot(x="wage_eur", y="short_name", data=dataYoung)
#Measured by value in Euro; age<=20
plt.xlabel("Wage in Euro")
plt.ylabel("Name of the Player")
plt.title("Young talents")
```

## Which countrie has the best players?

```python
data_2020Thebest20=fifa.sort_values('overall',ascending=False).head(20)
dataThebest20=data_2020Thebest20.groupby('nationality').count()
listindex2=list(dataThebest20.index.values)
plt.pie(dataThebest20.age,explode=None,autopct='%1.1f%%', labels=listindex2)
plt.axis('equal')
dataThebest20.index.values
plt.title("Country Comparison")
```

## Best Players in the world?

```python
dataTop10=fifa.sort_values('overall',ascending=False).head(10)
sns.set(style="whitegrid")
ax = sns.barplot(x="age", y="short_name", data=dataTop10)
plt.xlabel("Age of the Player")
plt.ylabel("Name of the Player")
plt.title("Best Playes in the Wolrd comparison")
```
